## Testing Deployed Model with Predictions

```python
import requests
import random

# Define the API endpoint
url = 'http://localhost:5000/predict'

# Possible values for categorical features based on your dataset
gender_options = ['Female', 'Male']
partner_options = ['Yes', 'No']
dependents_options = ['Yes', 'No']
phone_service_options = ['Yes', 'No']
multiple_lines_options = ['No phone service', 'Yes', 'No']
internet_service_options = ['DSL', 'Fiber optic', 'No']
online_security_options = ['Yes', 'No', 'No internet service']
online_backup_options = ['Yes', 'No', 'No internet service']
device_protection_options = ['Yes', 'No', 'No internet service']
tech_support_options = ['Yes', 'No', 'No internet service']
streaming_tv_options = ['Yes', 'No', 'No internet service']
streaming_movies_options = ['Yes', 'No', 'No internet service']
contract_options = ['Month-to-month', 'One year', 'Two year']
paperless_billing_options = ['Yes', 'No']
payment_method_options = ['Electronic check', 'Mailed check', 'Bank transfer (automatic)', 'Credit card (automatic)']

# Initialize counters for aggregate results
count_churn = 0  # Count for prediction = 1 (churn)
count_no_churn = 0  # Count for prediction = 0 (no churn)

# Generate 100 random test cases
for i in range(100):
    test_data = {
        'gender': random.choice(gender_options),
        'SeniorCitizen': random.randint(0, 1),
        'Partner': random.choice(partner_options),
        'Dependents': random.choice(dependents_options),
        'tenure': random.randint(1, 72),  # Assuming tenure ranges from 1 to 72 months
        'PhoneService': random.choice(phone_service_options),
        'MultipleLines': random.choice(multiple_lines_options),
        'InternetService': random.choice(internet_service_options),
        'OnlineSecurity': random.choice(online_security_options),
        'OnlineBackup': random.choice(online_backup_options),
        'DeviceProtection': random.choice(device_protection_options),
        'TechSupport': random.choice(tech_support_options),
        'StreamingTV': random.choice(streaming_tv_options),
        'StreamingMovies': random.choice(streaming_movies_options),
        'Contract': random.choice(contract_options),
        'PaperlessBilling': random.choice(paperless_billing_options),
        'PaymentMethod': random.choice(payment_method_options),
        'MonthlyCharges': round(random.uniform(18.25, 118.75), 2),  # Assuming MonthlyCharges ranges
        'TotalCharges': round(random.uniform(18.25, 8684.8), 2)  # Assuming TotalCharges ranges
    }

    # Send a POST request to the API
    response = requests.post(url, json=test_data)

    # Get the prediction from the response
    result = response.json().get('prediction')

    # Aggregate the results
    if result == 1:
        count_churn += 1
    elif result == 0:
        count_no_churn += 1

    # Print the prediction result for each test case
    print(f'Test Case {i+1}:', test_data)
    print('Prediction:', result)
    print('-' * 50)

# Print aggregate results
print(f"Total predicted to churn (1): {count_churn}")
print(f"Total predicted not to churn (0): {count_no_churn}")
```
