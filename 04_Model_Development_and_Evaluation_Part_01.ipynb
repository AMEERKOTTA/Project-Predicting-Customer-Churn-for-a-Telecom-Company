```python
## import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,\
            roc_auc_score, roc_curve, precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt
```

### Load and Prepare the Data

```python
## load the data
data = pd.read_csv('processed_customer_data.csv')
# data.info()

## split the data to features and target
features = data.drop(['Churn', 'customerID'], axis=1) 
target =  data['Churn']

print("Shape of the feature dataset :", features.shape)
print("Shape of the target dataset :", target.shape)
```

### Split the Data into Training and Testing Sets

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
print("Training Features : ", X_train.shape)
print("Testing Features : ", X_test.shape)
print("Training Target : ", y_train.shape)
print("Testing Target : ", y_test.shape)
```

### Training, Prediction and Evaluation

+ Accuracy: The proportion of correctly predicted instances.+ 
Precision, Recall, F1-Score: Provide deeper insights into the model's performance, especially for imbalanced datasets
+ 
Confusion Matrix: Shows the number of true positives, true negatives, false positives, and false negatives.

```python
# Initialize all models in a dictionary
models = {
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42),
    "XGBoost": XGBClassifier(n_estimators=100, learning_rate=0.1, random_state=42, eval_metric='logloss'),
    "SVM": SVC(kernel='rbf', random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes": GaussianNB(),
    "LightGBM": LGBMClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
}

# Loop through each model
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    y_pred_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None  # Check if model has predict_proba method
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name} Accuracy: {accuracy:.2f}")
    
    # Display classification report
    print(f"Classification Report - {model_name}:")
    print(classification_report(y_test, y_pred))
    
    # Display confusion matrix
    print(f"Confusion Matrix - {model_name}:")
    print(confusion_matrix(y_test, y_pred))
    
    # Plot and compute AUC-ROC Curve if model provides probability scores
    if y_pred_prob is not None:
        # AUC-ROC Score
        auc_score = roc_auc_score(y_test, y_pred_prob)
        print(f"AUC-ROC Score - {model_name}: {auc_score:.2f}")
        
        # ROC Curve
        fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
        plt.figure()
        plt.plot(fpr, tpr, color='darkorange', label=f'ROC curve (area = {auc_score:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'Receiver Operating Characteristic (ROC) Curve - {model_name}')
        plt.legend(loc="lower right")
        plt.show()
        
        # Precision-Recall Curve
        precision, recall, _ = precision_recall_curve(y_test, y_pred_prob)
        avg_precision = average_precision_score(y_test, y_pred_prob)
        print(f"Average Precision Score - {model_name}: {avg_precision:.2f}")
        
        plt.figure()
        plt.plot(recall, precision, color='b', label=f'Precision-Recall curve (area = {avg_precision:.2f})')
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title(f'Precision-Recall Curve - {model_name}')
        plt.legend(loc="lower left")
        plt.show()
    
    print("\n" + "="*50 + "\n")
```

### Evaluation Metrics and Their Interpretation

**Accuracy**

Accuracy measures the proportion of correctly predicted instances (both churn and non-churn) out of the total instances.

Best Performing Models by Accuracy:

+ Logistic Regression, Gradient Boosting, XGBoost, SVM: All achieved an accuracy of 81%.
+ LightGBM and Random Forest: Slightly lower with 80% accuracy.
+ Observation: Accuracy is a good starting point, but it might not be sufficient when dealing with imbalanced datasets (e.g., fewer churn cases).

**Precision**

Precision measures the proportion of true positive predictions (correctly predicted churn cases) out of all positive predictions. High precision indicates fewer false positives.

+ Precision for Class 1 (Churn):
+ Logistic Regression, XGBoost, Gradient Boosting, SVM: Precision around 0.67 - 0.69.
+ Naive Bayes: Lower precision at 0.45, indicating a higher rate of false positives.
+ Observation: A higher precision for churn prediction indicates that the model is less likely to wrongly classify non-churn customers as churn (fewer false alarms).


**Recall**

Recall (Sensitivity) measures the proportion of actual positive cases (churn cases) correctly predicted by the model. High recall indicates fewer false negatives.

+ Recall for Class 1 (Churn):
+ Naive Bayes: High recall at 0.87, indicating it captures most actual churn cases but at the cost of more false positives.
+ Logistic Regression, XGBoost, Gradient Boosting, SVM: Moderate recall around 0.51 - 0.60.
+ Observation: A higher recall means fewer churn cases are missed, which might be important if the cost of missing a churn is high.


**F1 Score**

F1 Score is the harmonic mean of precision and recall, providing a balance between them. It is useful when you need to balance false positives and false negatives.

+ F1 Score for Class 1 (Churn):
+ Logistic Regression, XGBoost, Gradient Boosting, SVM: Around 0.58 - 0.64.
+ Naive Bayes: Lower F1 score at 0.60, reflecting its imbalance between precision and recall.
+ Observation: F1 Score is a good metric when both precision and recall are important, and you want a balance between false positives and false negatives.

**AUC-ROC Score**

AUC-ROC (Area Under the Receiver Operating Characteristic Curve) measures the model's ability to distinguish between classes (churn vs. non-churn). The higher the AUC, the better the model distinguishes between the positive and negative classes.

+ Best Performing Models by AUC-ROC:
+ Gradient Boosting, XGBoost, LightGBM: Usually perform well with high AUC scores.
+ Observation: AUC-ROC is particularly useful when dealing with imbalanced datasets and when you want to understand the model's ability to discriminate between the two classes.


**Precision-Recall Curve**

Precision-Recall Curve helps understand the trade-off between precision and recall for different thresholds. The area under this curve (Average Precision Score) provides insight into model performance, especially in imbalanced datasets.

+ Observation: Look for models with a high area under the precision-recall curve, which indicates better performance in predicting the minority class.

**Inference and Recommendations**

**Best Models for Churn Prediction**

+ Logistic Regression, XGBoost, Gradient Boosting, SVM: All of these models have similar accuracy (around 81%) and balanced performance in terms of precision, recall, and F1-score.
+ XGBoost and Gradient Boosting: Generally considered strong models for classification tasks due to their ability to handle non-linear relationships and manage overfitting through boosting techniques.
+ LightGBM: Also a good option, especially for large datasets and when computation speed is a concern. It is known for high performance and efficiency.

**Models to Improve or Avoid**

+ Decision Tree, KNN, Naive Bayes: These models show relatively lower performance, with lower accuracy, F1-scores, and poorer trade-offs between precision and recall. Naive Bayes, for example, has high recall but low precision, leading to many false positives.

**Evaluation Strategy Going Forward**

+ Focus on models that provide a good balance between precision and recall (e.g., Logistic Regression, XGBoost, Gradient Boosting, LightGBM).
+ Use AUC-ROC and Precision-Recall Curves to understand the discriminative ability of each model, especially important in imbalanced datasets.

**Next Steps**

+ Hyperparameter Tuning: Fine-tune the hyperparameters of the top-performing models (e.g., XGBoost, Gradient Boosting, LightGBM) using techniques like Grid Search or Random Search to further improve performance.
+ Handle Class Imbalance: Consider techniques like SMOTE (Synthetic Minority Over-sampling Technique) or class weighting to address class imbalance and improve model performance.
+ Ensemble Methods: Combine multiple models using ensemble methods (like stacking or bagging) to leverage the strengths of each and improve overall performance.
